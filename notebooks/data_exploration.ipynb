```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
import seaborn as sns
```

## Load prepocessing data

```python
sales=pd.read_csv('../data/prepocessing/sales.csv')
sales_and_purchase_prices=pd.read_csv('../data/prepocessing/sales_and_purchase_prices.csv')
```

## Analysis top brands with the highest benifice

**Marge benificaire:
Marge bénéficiaire totale = (Prix de vente - Prix d'achat) * Quantité vendue

```python
sales_and_purchase_prices['Profit']=(sales_and_purchase_prices['SalesPrice']-sales_and_purchase_prices['PurchasePrice'])*sales_and_purchase_prices['SalesQuantity']
```

```python
sales_and_purchase_prices.head()
```

```python
df_top_brand_profit = sales_and_purchase_prices.groupby(['Description'])['Profit'].sum().sort_values(ascending=False)[:10]
```

```python
import pandas as pd
from matplotlib import pyplot as plt


name = list(df_top_brand_profit.index)
price = list(df_top_brand_profit.values)

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(name, price)

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
	ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

# Add x, y gridlines
ax.grid(True, color='grey', linestyle='-.', linewidth=0.5, alpha=0.2)


# Show top values
ax.invert_yaxis()

# Add annotation to bars
for i in ax.patches:
	plt.text(i.get_width()+0.2, i.get_y()+0.5,
			str(round((i.get_width()), 2)),
			fontsize = 10, fontweight ='bold',
			color ='grey')

# Add Plot Title
ax.set_title('Top 10 Brand with the highest benifice',
			loc ='left', )



# Add Text watermark
fig.text(0.9, 0.15, 'AbdoulNasserProd', fontsize = 12,
		color ='grey', ha ='right', va ='bottom',
		alpha = 0.7)


# Adding Xticks
plt.xlabel('Dollards', fontweight ='bold', fontsize = 15)
plt.ylabel('Brand', fontweight ='bold', fontsize = 15)

# Show Plot

plt.show()
```

**trend analysis for best-selling products**

```python
sales.head()
```

```python
sales['SalesDate'] = pd.to_datetime(sales['SalesDate'])
sales.sort_values('SalesDate').reset_index(drop=True)
df_grouped_by_brand=sales.groupby(['Description']).agg({
    'SalesQuantity': 'sum',
    'Amount': 'sum'})
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_grouped_by_brand is your DataFrame with grouped data

# Get the top 5 products based on SalesQuantity
top_products = df_grouped_by_brand.nlargest(5, 'SalesQuantity')

# Get the descriptions of the top 5 products
top_product_descriptions = top_products.index

# Filter the sales data to include only the top products
top_product_sales = sales[sales['Description'].isin(top_product_descriptions)]

# Group the filtered sales data by Description and Date
top_product_sales_grouped = top_product_sales.groupby(['Description', 'SalesDate']).agg({
    'SalesQuantity': 'sum'
}).reset_index()

# Create a line plot for each top product's sales evolution over time
plt.figure(figsize=(10, 6))

for description in top_product_descriptions:
    product_data = top_product_sales_grouped[top_product_sales_grouped['Description'] == description]
    plt.plot(product_data['SalesDate'], product_data['SalesQuantity'], label=description)

plt.xlabel('Date')
plt.ylabel('Sales Quantity')
plt.title('Sales Evolution for Top 5 Products')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

# Top sellers quantity/Amount

```python
 df_grouped_by_vendor=sales.groupby(['VendorName']).agg({
    'SalesQuantity': 'sum',
    'Amount': 'sum'})
```

```python
df_grouped_by_vendor=df_grouped_by_vendor.sort_values(by='SalesQuantity',ascending=False).reset_index()
```

```python
df_grouped_by_vendor
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Assuming you have a dataframe df_grouped_by_vendor with 'VendorName', 'SalesQuantity', and 'Amount' columns
df_grouped_by_vendor = df_grouped_by_vendor.sort_values(by='SalesQuantity', ascending=False).reset_index()

# Select the top 5 vendors
top_5_vendors = df_grouped_by_vendor.head(5)

# Extract vendor names, sales quantities, and amounts
vendor_names = top_5_vendors['VendorName']
sales_quantities = top_5_vendors['SalesQuantity']
amounts = top_5_vendors['Amount']

# Convert large numbers to abbreviations (e.g., 1000 -> 1K, 1000000 -> 1M, etc.)
def format_abbreviations(x):
    if x >= 1e6:
        return f'{x/1e6:.1f}M'
    elif x >= 1e3:
        return f'{x/1e3:.1f}K'
    else:
        return f'{x:.0f}'

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6), sharey=True)

# Plot the sales quantities on the left side of the x-axis
ax1.bar(vendor_names, sales_quantities, color='skyblue', edgecolor='black')
ax1.set_xlabel('Vendor Name')
ax1.set_ylabel('Quantity')
ax1.set_title('Top 5 Vendors - Sales Quantity')

# Add values on top of the bars with abbreviations
for i, v in enumerate(sales_quantities):
    ax1.text(i, v, format_abbreviations(v), ha='center', va='bottom', fontweight='bold')

# Format the y-axis tick labels as long numbers
ax1.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Create a twin y-axis for the second subplot
ax2y = ax2.twinx()

# Plot the sales amounts on the right side of the x-axis
ax2y.bar(vendor_names, amounts, color='orange', edgecolor='black')
ax2y.set_ylabel('Amount')
ax2.set_xlabel('Vendor Name')
ax2.set_title('Top 5 Vendors - Sales Amount')

# Add values on top of the bars with abbreviations
for i, v in enumerate(amounts):
    ax2y.text(i, v, format_abbreviations(v), ha='center', va='bottom', fontweight='bold')

# Format the twin y-axis tick labels as long numbers
ax2y.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Rotate the x-axis tick labels
plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')
plt.setp(ax2.get_xticklabels(), rotation=45, ha='right')

# Add gridlines to both subplots
ax1.grid(axis='y', linestyle='dashed', alpha=0.5)
ax2y.grid(axis='y', linestyle='dashed', alpha=0.5)

# Adjust spacing between subplots
plt.tight_layout()

plt.show()
```
